```python
import csv
```

```python
sentences=[]
with open('sentence.txt','r', encoding='UTF8') as f:
    reader = csv.reader(f)
    
    for r in reader:
        sentences.append(*r)
```

```python
word_count=dict()

for s in sentences:
    words = s.split()
    word_cnt = len(words)
    
    for i in range(word_cnt):
        if i > 2:
            if i < word_cnt-2:
                before1 = str(words[i]+" "+words[i-1])
                before2 = str(words[i]+" "+words[i-2])
                after1 = str(words[i]+" "+words[i+1])
                after2 = str(words[i]+" "+words[i+2])

                # 전 1번째
                if before1 not in word_count:
                    word_count[before1] = 1
                else:
                    word_count[before1] = word_count[before1]+1

                # 전 2번째
                if before2 not in word_count:
                    word_count[before2] = 1
                else:
                    word_count[before2] = word_count[before2]+1

                # 뒤 1번째
                if after1 not in word_count:
                    word_count[after1] = 1
                else:
                    word_count[after1] = word_count[after1]+1

                # 뒤 2번째
                if after2 not in word_count:
                    word_count[after2] = 1
                else:
                    word_count[after2] = word_count[after2]+1
                    
            elif i < word_cnt-1:
                before1 = str(words[i]+" "+words[i-1])
                before2 = str(words[i]+" "+words[i-2])
                after1 = str(words[i]+" "+words[i+1])

                # 전 1번째
                if before1 not in word_count:
                    word_count[before1] = 1
                else:
                    word_count[before1] = word_count[before1]+1

                # 전 2번째
                if before2 not in word_count:
                    word_count[before2] = 1
                else:
                    word_count[before2] = word_count[before2]+1

                # 뒤 1번째
                if after1 not in word_count:
                    word_count[after1] = 1
                else:
                    word_count[after1] = word_count[after1]+1

        if i > 1:
            if i < word_cnt-2:
                before1 = str(words[i]+" "+words[i-1])
                after1 = str(words[i]+" "+words[i+1])
                after2 = str(words[i]+" "+words[i+2])

                # 전 1번째
                if before1 not in word_count:
                    word_count[before1] = 1
                else:
                    word_count[before1] = word_count[before1]+1

                # 뒤 1번째
                if after1 not in word_count:
                    word_count[after1] = 1
                else:
                    word_count[after1] = word_count[after1]+1

                # 뒤 2번째
                if after2 not in word_count:
                    word_count[after2] = 1
                else:
                    word_count[after2] = word_count[after2]+1
                    
            elif i < word_cnt-1:
                before1 = str(words[i]+" "+words[i-1])
                after1 = str(words[i]+" "+words[i+1])

                # 전 1번째
                if before1 not in word_count:
                    word_count[before1] = 1
                else:
                    word_count[before1] = word_count[before1]+1

                # 뒤 1번째
                if after1 not in word_count:
                    word_count[after1] = 1
                else:
                    word_count[after1] = word_count[after1]+1
```

```python
word_key = list(word_count.keys())
word_count_list=[]

for k in word_key:
    if word_count[k] >= 10:
        word_count_list.append([k,word_count[k]])

len(word_count_list)
```

```python
csvfile = open("neighboring_word_count.csv", "w", newline="", encoding='UTF8')

csvwriter = csv.writer(csvfile)
for row in word_count_list:
    csvwriter.writerow(row)

csvfile.close()
```
